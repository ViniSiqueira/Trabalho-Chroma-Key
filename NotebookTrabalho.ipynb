```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import median
```

```python
#pasta = 'Rainha'
#pasta = 'Formas/'
#pasta = 'Corvos/'
pasta = 'Corvo/'
```

```python
#imagem = 'rainha'
#imagem = 'formas'
#imagem = 'corvos'
imagem = 'corvo'
```

```python
imgOriginal = cv2.imread(f'/content/gdrive/MyDrive/Visão Computacional/Trabalho1/Imagens/{pasta}/{imagem}.bmp')
background = cv2.imread('/content/gdrive/MyDrive/Visão Computacional/Trabalho1/Imagens/background.bmp')
```

```python
imgOriginalAltura, imgOriginalLargura = imgOriginal.shape[:2]
backgroundResize = cv2.resize(background, (imgOriginalLargura, imgOriginalAltura))
```

```python
hsv = cv2.cvtColor(imgOriginal, cv2.COLOR_BGR2HSV)
backgroundHsv = cv2.cvtColor(backgroundResize, cv2.COLOR_BGR2HSV)
```

```python
lower_green = np.array([50, 100, 100])
upper_green = np.array([70, 255, 255])
```

```python
_, mask = cv2.threshold(cv2.inRange(hsv, lower_green, upper_green), 0, 255, cv2.THRESH_BINARY)
save = cv2.imwrite(f'/content/gdrive/MyDrive/Visão Computacional/Trabalho1/Imagens/{pasta}/{imagem}Binarizada.bmp',mask)
```

```python
blur = cv2.GaussianBlur(mask, (11,11), 0)
save = cv2.imwrite(f'/content/gdrive/MyDrive/Visão Computacional/Trabalho1/Imagens/{pasta}/{imagem}Gaussiano.bmp',blur)
```

```python
sobelx = cv2.Sobel(blur, cv2.CV_64F, 1, 0, (11,11))
sobely = cv2.Sobel(blur, cv2.CV_64F, 0, 1, (11,11))
mascara_sobel = cv2.convertScaleAbs(cv2.addWeighted(sobelx, 0.5, sobely, 0.5, 0))
save = cv2.imwrite(f'/content/gdrive/MyDrive/Visão Computacional/Trabalho1/Imagens/{pasta}/{imagem}Sobel.bmp',mascara_sobel)
```

```python
maskedBackground = cv2.bitwise_and(backgroundHsv, backgroundHsv, mask=mask)

maskInvertida = cv2.bitwise_not(mask)

maskedImg = cv2.bitwise_and(hsv, hsv, mask=maskInvertida)

resultHsv = cv2.bitwise_or(maskedBackground, maskedImg)

resultBackground = cv2.cvtColor(resultHsv, cv2.COLOR_HSV2BGR)

save = cv2.imwrite(f'/content/gdrive/MyDrive/Visão Computacional/Trabalho1/Imagens/{pasta}/{imagem}Final.bmp',resultBackground)
```
